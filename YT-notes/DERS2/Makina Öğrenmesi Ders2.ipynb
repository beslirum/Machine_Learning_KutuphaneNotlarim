```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
# formülümüz : y = mx + b (Y bağımlı, X bağımsız değişkenimizdi. Bu formül tek değişken için kullanılıyor.)
```

```python
df = pd.read_csv("student.csv")
```

```python
df.head(3)
```

```python
'''
buraya bakarsak 2 tane x değerimiz var. ilk x kaçıncı sınıf olduğunu,ikinci x çalışma süresini gösterir. Ama bu iki değişkenin kat sayısını yani ne kadar etkileyeceğini (a1 ve a2)
bilmiyoruz. Sabit değer olan b'yi de bilmiyoruz
'''

# y yani tahmin etmek istediğimiz değer ise marks olur
```

```python
y = df[["Marks"]]
x = df[["number_courses", "time_study"]]
```

```python
df.info()
```

```python
linearModel = LinearRegression()          #Eğer en başta kütüphanede direkt bunu import etmeseydik burada eşitliğimiz:

#sklearn.linear_model.LinearRegression() şeklinde gösterilecekti

model = linearModel.fit(x,y)
```

```python
model.predict([[4, 4]])     #predict yapısı ile tahmin ediyoruz.
```

```python
df['Marks'].max()     #max notu gösterelim
```

```python
model.score(x,y)    #başarı orani
```
