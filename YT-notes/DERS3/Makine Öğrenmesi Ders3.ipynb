```python
from sklearn.linear_model import  LinearRegression
import pandas as pd
```

```python
df = pd.read_csv('Student_Marks.csv')
```

```python
df.head(3)
```

```python
df.columns=['sinif', 'saat', 'puan']
```

```python
df.head(3)
```

```python
y = df[['puan']]
x=df[['sinif', 'saat']]
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
# y = ax + b
```

```python
model.coef_     #HER BİR SABİTİN KAT SAYISINI VERDİ
```

```python
model.intercept_      #SABİT DEĞERİ (B) VERİR
```

```python
model.predict([[4, 5]])   #TAHMİN YAPALİM SİMDİ
```
