```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.head(3)
```

Öncelikle kategorisel verileri Dummy Methods ile nümerik veriye dönüştürmeliyiz

```python
df = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], drop_first=True)
```

```python
df.head(3)
```

```python
y = df[['charges']]
x = df.drop('charges', axis=1)
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
model.score(x,y)
```

```python
model.predict([[19, 26, 0, 1,1,0,0,1]])
```

```python
df_hata = pd.DataFrame()
```

```python
df_hata['gercek']=y
```

```python
y_tahmin = model.predict(x)
```

```python
df_hata['tahmin'] = y_tahmin
```

```python
df_hata
```

```python
df_hata['hata orani'] = (df_hata['gercek'])-(df_hata['tahmin'])
```

```python
df_hata
```


**MSE (Mean Squared Error - Ortalama Kare Hatası):**

*   MSE, tahmin edilen değerler ile gerçek değerler arasındaki farkların karelerinin ortalamasını hesaplar.
*   MSE'nin küçük olması, modelin daha iyi performans gösterdiğini gösterir.

*   Ancak, MSE'nin bir dezavantajı, ölçü birimlerinin karesi olmasıdır, bu nedenle orijinal verilere göre yorumlaması zor olabilir.

*   Neden kare alırız peki? Yukarıdaki tabloda görüldüğü gibi negatif veya pozitif değer alabilir. Bunları mutlak olarak belirtmemiz işlemlerin sıfırlanmasını engelleyecek. Kare alma da bu devrede işimize yarayacak.


```python
df_hata['MSE'] = df_hata['hata orani']**2
```

**MAE (Mean Absolute Error - Ortalama Mutlak Hata):**

* MAE, tahmin edilen değerler ile gerçek değerler arasındaki mutlak farkların ortalamasını hesaplar.
* MAE, hataların mutlak değerlerini kullanır, bu nedenle büyük aykırı değerlerin etkisi daha azdır.
* MAE'nin küçük olması, modelin daha iyi performans gösterdiğini gösterir.

```python
df_hata['MAE']=(np.abs((df_hata['hata orani']) - (df_hata['gercek'])))/2
```

**MAPE (Mean Absolute Percentage Error - Ortalama Mutlak Yüzde Hatası):**

* MAPE, tahmin edilen ve gerçek değerler arasındaki hataların yüzde olarak ortalama mutlak değerini hesaplar.
* MAPE, hataların orijinal verilere göre yüzde olarak yorumlanmasını sağlar.
* MAPE'nin küçük olması, modelin daha iyi performans gösterdiğini gösterir.
* Ancak, MAPE'nin sıfır olması durumunda hata bölü sıfır hatası (divided by zero error) alınabilir.

```python
df_hata['MAPE']=np.abs(((df_hata['gercek'])- df_hata['tahmin']/df_hata['gercek']) * 100)
```

Mean() ile değerleirn ortalamasını döndürürüz.

```python
df_hata.mean()
```
