```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Audi_A1_listings.csv')
```

```python
df.head(3)      #PPY=Price Per Year demek. Bu TAHMİN için verilebilir, ama hazır fiyatıverilmemeli çünkü biz bilmediğimiz şeyleri hesaplamaya çalışacağız.
#href linkidir, bunun gibi değerler etkilemez tahminimizi
#Number_of_Owners, yani aracın kaçıncı sahibi olduğu fiyatı etkiler
#index sıralaması ETKİLEMEZ
```

```python
# y = ax + b
#Kullanacağımız değerleri alalim, kullanmayacağımız değerleri atalım. Bunu da pandasın DROP metodu ile yaparız.
df=df.drop(columns=['index', 'href', 'MileageRank', 'PriceRank', 'PPYRank', 'Score'])

#Daha önceden sildiğim için tekrar sildiğimde error geldi, SYNTAX DOĞRU
```

```python
#Preprocessing işlemi yaptık bu şekilde. Bu işlem; veri temizliği, dönüştürülmesi ve hazır hale getirilmesini ifade eder!!
```

```python
#y = ax + b
#String ifadeleri 'a' gibi bir katsayı ile çarpamayız. Bu gibi verilere makine öğrenmesinde KATEGORİSEL VERİ deriz.
#Kat sayı ile çarpılabilen verilere NÜMERİK VERİ deriz.

#Engine değeri nümerik bir veri, ama sonunda 'L' harfi var. Bunu kaldırmamız lazım
df.info()
```

```python
df.head(3)
```

```python
df.columns=['yil', 'kasa', 'mil', 'motor', 'ps', 'vites', 'yakit', 'sahip', 'fiyat', 'ppy']
```

```python
df.head(3)
```

```python
df['motor'] = df['motor'].str.replace("L", "")  #Bu bunu hala nümerik bir veri yapmadı
```

```python
df['motor'] = pd.to_numeric(df['motor']) #şimdi nümerik oldu
```

```python
df.info()
```

```python
df.head(3)
```

```python
#Hala KATEGORİSEL verilerim var
#Bu durumda da Pandasın DUMMY Metodunu kullanarak bunları nümerik veriye çevireceğiz
df = pd.get_dummies(df,columns=['kasa', 'vites', 'yakit'],drop_first=True)     #ilk secenegi kaldirdik son kısım ile
```

```python
df.head(3)
```

```python
y = df[['fiyat']]
x=df.drop("fiyat", axis=1)      #axis=1 sütunlardan fiyatı sil demek
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.predict([[2017, 30000, 1.6,110, 1, 2600, 0 , 1]])
```

```python
model.score(x,y)
```
